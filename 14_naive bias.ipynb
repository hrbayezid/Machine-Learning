```python
import pandas as pd
import kagglehub as kh
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB , BernoulliNB , MultinomialNB
from mlxtend.plotting import plot_decision_regions
```

```python
path = kh.dataset_download("somya2115/placement-csv")
database = pd.read_csv(f"{path}/placement.csv")
database = database.drop("Unnamed: 0", axis=1)
database.head()
```

```python
x = database.drop("placement",axis=1)
y =  database["placement"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.25,random_state=42)
```

```python
gn = GaussianNB()
mn = MultinomialNB()
bl = BernoulliNB()
```

Gaussian

```python
sns.kdeplot(database["cgpa"])
plt.show()
```

```python
sns.kdeplot(database["iq"])
plt.show()
```

```python
gn.fit(x_train,y_train)
gn.score(x_test,y_test)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gn)
plt.show()
```

Multinomial

```python
mn.fit(x_train,y_train)
mn.score(x_test,y_test)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mn)
plt.show()
```

```python
bl.fit(x_train,y_train)
bl.score(x_test,y_test)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bl)
plt.show()
```

# Naive Bayes Classifier: Simple Explanation

## What is Naive Bayes?
Naive Bayes is a **machine learning algorithm** used for **classification** tasks. It is based on **Bayes' Theorem** and is called "naive" because it assumes that all features are **independent** of each other.

---

## Key Points

- **Supervised Learning**: Needs labeled data to train.
- **Probabilistic Model**: Predicts the probability of each class.
- **Fast & Simple**: Works well with small datasets and is easy to implement.
- **Common Uses**: Spam detection, sentiment analysis, document classification.

---

## How Does It Work?

A. **Bayes' Theorem**  
    -Expressed as:  
$$
P(A \mid B) = \frac{P(B \mid A) \times P(A)}{P(B)}
$$
 
2. **Naive Assumption**  
    - Thinks all features are separate and don't affect each other.
    - Example: In spam detection, "free" and "win" are treated as unrelated.

3. **Classification**  
    - Calculates the chance for each class.
    - Picks the class with the **highest chance**.

---

## Types of Naive Bayes

- **Gaussian Naive Bayes**: For continuous data (assumes normal distribution).
- **Multinomial Naive Bayes**: For count data (like word counts in text).
- **Bernoulli Naive Bayes**: For binary/boolean features.

---

## Advantages

- **Simple and fast**
- **Works well with high-dimensional data**
- **Performs well with small datasets**

## Limitations

- **Assumes feature independence** (not always true in real life)
- **Not suitable for complex relationships between features**

---
