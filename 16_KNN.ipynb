```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub as kh

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
from mlxtend.plotting import plot_decision_regions
```

KNN Classification

```python
path = kh.dataset_download("rakeshrau/social-network-ads")
dataset = pd.read_csv(f"{path}/Social_Network_Ads.csv")

dataset.drop(["User ID","Gender"],axis=1, inplace=True)
dataset.head()
```

```python
sns.scatterplot(x="Age",y= "EstimatedSalary",data= dataset,hue="Purchased")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Purchased"]

sc =StandardScaler()
x= pd.DataFrame(sc.fit_transform(x),columns=x.columns)

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2,random_state=42)
kn = KNeighborsClassifier(n_neighbors=3)
kn.fit(x_train,y_train)
print( kn.score(x_train,y_train)*100, kn.score(x_test,y_test)*100)
```

```python
kn.predict([[35,20000]])
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=kn)
plt.show()
```
