```python
import pandas as pd
import kagglehub as kh
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC , SVR
from mlxtend.plotting import plot_decision_regions
```

```python
data_path = kh.dataset_download("somya2115/placement-csv")

dataset = pd.read_csv(f"{data_path}/placement.csv")
dataset.drop(columns="Unnamed: 0",axis=1,inplace=True)
dataset.head()
```

```python
sns.scatterplot(x="cgpa" , y ="iq", data=dataset ,hue="placement")
```

```python
x = dataset[["cgpa","iq"]]
y = dataset ["placement"]


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2,random_state=4)
sv = SVC( kernel = "linear") #kernel = linear', 'poly', 'rbf', 'sigmoid', 'precomputed'
sv.fit(x_train,y_train)
print(sv.score(x_train,y_train)*100,sv.score(x_test,y_test)*100)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```

```python
path = kh.dataset_download("thembatman/make-moons-from-sklearn")

dataset = pd.read_csv(f"{path}/make_moons.csv")
dataset.head()
```

```python
sns.scatterplot(x ="X", y= "Y",data=dataset,hue="label")
```

```python
x = dataset[["X","Y"]]
y = dataset ["label"]


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2,random_state=1)
svr = SVR( kernel = "rbf") #kernel = linear', 'poly', 'rbf', 'sigmoid', 'precomputed'
svr.fit(x_train,y_train)
print(svr.score(x_train,y_train)*100,svr.score(x_test,y_test)*100)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=svr)
plt.show()
```
