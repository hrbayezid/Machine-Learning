```python
import pandas as pd
import kagglehub as kh
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, LeaveOneOut,LeavePOut,KFold,StratifiedKFold
```

```python
data_path = kh.dataset_download("tann22u/placement-dataset")

dataset = pd.read_csv(f"{data_path}/placement.csv")
dataset
```

```python
x = dataset.iloc[:,:-1]
y = dataset["package"]

# crv = cross_val_score(LinearRegression(),x,y,cv =5)
crv = cross_val_score(estimator=LinearRegression(),X= x,y= y,cv=KFold(n_splits=20))
crv.sort()
crv*100
```

```python
new_data = dataset.head(10)
x_new = new_data.iloc[:,:-1]
y_new = new_data["package"]


lo = LeaveOneOut ()

for train , test in lo.split(x_new,y_new):
    print( train,test)
```

```python
lp = LeavePOut(p=5)

for train,test in lp.split(x_new,y_new):
    print(train,test)
```

```python
kf = KFold(n_splits=5)

for train,test in kf.split(x_new,y_new):
    print(train,test)
```

```python
kf = KFold(n_splits=5)

for train,test in kf.split(x_new,y_new):
    print(train,test)
```
