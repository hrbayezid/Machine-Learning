***HAC***
- it can be useed if dataset is  linearly separable,

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
import scipy.cluster.hierarchy as sc
from sklearn.cluster import AgglomerativeClustering
```

```python
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True).frame

iris.drop("target",axis=1, inplace=True)
iris.head()
```

```python
sns.pairplot(data=iris)
plt.show()
```

```python
sc.dendrogram(sc.linkage(iris,"complete","euclidean"))
plt.show()
```

```python
ag = AgglomerativeClustering(n_clusters=2,linkage="complete")

iris["predict"] = ag.fit_predict(iris)
iris.head()
```

```python
sns.pairplot(data=iris, hue= "predict")
plt.show()
```
