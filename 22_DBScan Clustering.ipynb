```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
x,y = make_moons(n_samples= 200 , noise= .1)
```

```python
dict = {"data": x[:,0], "data2": x[:,1] }

df = pd.DataFrame(dict)
df
```

```python


sns.scatterplot(x = "data" ,y = "data2" , data=df)
```

```python
db = DBSCAN(eps= .2, min_samples=5 , metric= "euclidean")

df["predict"] = db.fit_predict(x)
```

```python
sns.scatterplot(x = "data" ,y = "data2" , data=df, hue = "predict")
```
