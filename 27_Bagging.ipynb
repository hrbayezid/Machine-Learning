```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
from sklearn.datasets import make_moons
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

bagging meta estimator

```python
x,y =  make_moons(n_samples= 1000 , noise=.2)

data = {"x1": x[:,0],"x2":x[:,1] , "y":y}

df = pd.DataFrame(data)

```

```python
sns.scatterplot(x= "x1", y="x2" , data=df,  hue= "y")
plt.show()
```

```python
xx = df.iloc[:,:-1]
yy = df["y"]

x_train, x_test ,y_train, y_test = train_test_split(xx,yy, test_size=.2, random_state=42)
```

```python
bg = BaggingClassifier(estimator= SVC(), n_estimators=20)

bg.fit(x_train,y_train)
```

```python
print(bg.score(x_train,y_train)*100,bg.score(x_test,y_test)*100)
```

Random forest

```python
rf = RandomForestClassifier(n_estimators=100,max_depth=7)
rf.fit(x_train,y_train)

print(rf.score(x_train,y_train)*100,rf.score(x_test,y_test)*100)
```
