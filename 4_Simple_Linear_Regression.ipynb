```python
x = dataset[["cgpa"]]
y = dataset [["package"]]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.25,random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)

```

```python
lr.predict([[6.89]])
```

```python
lr.score(x_test,y_test)
```

# Simple Linear Regression

Simple Linear Regression is a statistical method used to model the relationship between two continuous variables: one independent variable (predictor) and one dependent variable (response).

## Key Concepts

- **Equation:**  
    The relationship is modeled as:  
    $$ y = mx + c $$
    where:
    - \( y \): Dependent variable (target)
    - \( x \): Independent variable (feature)
    - \( m \): Slope of the line (coefficient)
    - \( c \): Intercept (constant)

- **Goal:**  
    To find the best-fitting straight line through the data points that minimizes the sum of squared differences between observed and predicted values.

- **Assumptions:**
    - Linearity: The relationship between \( x \) and \( y \) is linear.
    - Independence: Observations are independent of each other.
    - Homoscedasticity: Constant variance of errors.
    - Normality: Errors are normally distributed.

## Steps in Simple Linear Regression

1. **Visualize the Data:**  
     Use scatter plots to check for a linear relationship.

2. **Split the Data:**  
     Divide the dataset into training and testing sets.

3. **Fit the Model:**  
     Use the training data to estimate the slope and intercept.

4. **Make Predictions:**  
     Predict the dependent variable for new or test data.

5. **Evaluate the Model:**  
     Use metrics like R-squared and Mean Squared Error to assess performance.

## Applications

- Predicting outcomes (e.g., salary based on experience, house price based on size)
- Understanding relationships between variables
