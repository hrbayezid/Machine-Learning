# Notes on Cost Function

A **cost function** is a way to measure how well a machine learning model is performing.

---

## What is a Cost Function?

- It is a mathematical formula.
- It tells us **how far off** our model's predictions are from the actual values.
- The goal is to **minimize** the cost function (make the error as small as possible).

---

## Why is it Important?

- Helps us know if our model is learning correctly.
- Guides the training process to improve predictions.

---

## Simple Example

Suppose you predict house prices.  
If the actual price is $200,000 and your model predicts $210,000:

- The error is $10,000.
- The cost function calculates this error for all predictions.

---

### Types of Cost Functions

#### For Regression

- **Mean Squared Error (MSE):**  
    Measures the average squared difference between actual and predicted values.  
    **Equation:**  
    $$
    \mathrm{MSE} = \frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2
    $$

- **Mean Absolute Error (MAE):**  
    Measures the average absolute difference between actual and predicted values.  
    **Equation:**  
    $$
    \mathrm{MAE} = \frac{1}{n} \sum_{i=1}^n |y_i - \hat{y}_i|
    $$

- **Root Mean Squared Error (RMSE):**  
    The square root of MSE, gives error in the same units as the target variable.  
    **Equation:**  
    $$
    \mathrm{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^n (y_i - \hat{y}_i)^2}
    $$

#### For Classification
- **Binary Cross-Entropy (BCE):**  
    Used for binary classification problems.  
    **Equation:**  
    $$
    \mathrm{BCE} = -\frac{1}{n} \sum_{i=1}^n \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
    $$

- **Categorical Cross-Entropy (CCE):**  
    Used for multiclass classification problems.  
    **Equation:**  
    $$
    \mathrm{CCE} = -\sum_{i=1}^n \sum_{j=1}^C y_{i,j} \log(\hat{y}_{i,j})
    $$
